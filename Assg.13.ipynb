## 1. Crop Area Distribution
- Analysis: Calculate the total area allocated to rice, wheat, and maize.
- Visualization: Bar Chart
- How: Use a bar chart to display total areas for each crop. Each bar represents a crop, and the height reflects the area in 1000 ha.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ICRISAT-District Level Data.csv')
```

```python
# Total area allocated to Rice, Wheat and Maize:
x = df['RICE AREA (1000 ha)'].sum() + df['WHEAT AREA (1000 ha)'].sum() + df['MAIZE AREA (1000 ha)'].sum()
print(f"The total area allocated to rice, wheat, and maize is : {round(x)} per 1000 ha.")
```

```python
# use a bar chart to display total areas for each crop. 
# Each bar represents a crop, and the height reflects the area in 1000 ha.

crop_totals = df.drop(columns= ['Dist Code', 'Year', 'State Code', 'State Name', 'Dist Name',
                               'RICE PRODUCTION (1000 tons)','RICE YIELD (Kg per ha)',
                               'WHEAT PRODUCTION (1000 tons)', 'WHEAT YIELD (Kg per ha)','KHARIF SORGHUM PRODUCTION (1000 tons)',
       'KHARIF SORGHUM YIELD (Kg per ha)', 'RABI SORGHUM PRODUCTION (1000 tons)', 'RABI SORGHUM YIELD (Kg per ha)', 'SORGHUM PRODUCTION (1000 tons)',
       'SORGHUM YIELD (Kg per ha)', 'SORGHUM PRODUCTION (1000 tons)', 'SORGHUM YIELD (Kg per ha)', 'PEARL MILLET PRODUCTION (1000 tons)', 'PEARL MILLET YIELD (Kg per ha)',   
'MAIZE PRODUCTION (1000 tons)',
       'MAIZE YIELD (Kg per ha)',
       'FINGER MILLET PRODUCTION (1000 tons)',
       'FINGER MILLET YIELD (Kg per ha)', 
       'BARLEY PRODUCTION (1000 tons)', 'BARLEY YIELD (Kg per ha)',
        'CHICKPEA PRODUCTION (1000 tons)',
       'CHICKPEA YIELD (Kg per ha)', 
       'PIGEONPEA PRODUCTION (1000 tons)', 'PIGEONPEA YIELD (Kg per ha)',
        'MINOR PULSES PRODUCTION (1000 tons)',
       'MINOR PULSES YIELD (Kg per ha)', 
       'GROUNDNUT PRODUCTION (1000 tons)', 'GROUNDNUT YIELD (Kg per ha)',
        'SESAMUM PRODUCTION (1000 tons)',
       'SESAMUM YIELD (Kg per ha)', 
       'RAPESEED AND MUSTARD PRODUCTION (1000 tons)',
       'RAPESEED AND MUSTARD YIELD (Kg per ha)', 
       'SAFFLOWER PRODUCTION (1000 tons)', 'SAFFLOWER YIELD (Kg per ha)',
       'CASTOR PRODUCTION (1000 tons)',
       'CASTOR YIELD (Kg per ha)', 
       'LINSEED PRODUCTION (1000 tons)', 'LINSEED YIELD (Kg per ha)',
       'SUNFLOWER PRODUCTION (1000 tons)',
       'SUNFLOWER YIELD (Kg per ha)', 
       'SOYABEAN PRODUCTION (1000 tons)', 'SOYABEAN YIELD (Kg per ha)',
        'OILSEEDS PRODUCTION (1000 tons)',
       'OILSEEDS YIELD (Kg per ha)', 
       'SUGARCANE PRODUCTION (1000 tons)', 'SUGARCANE YIELD (Kg per ha)',
        'COTTON PRODUCTION (1000 tons)',
       'COTTON YIELD (Kg per ha)']).sum()
crop_totals
```

```python
crop_totals.plot(kind = 'bar' , figsize=(15,8))
plt.xticks(rotation = 45, ha = 'right')
plt.title('Bar chart displaying total areas for each crop')
plt.xlabel('Crop Variety')
plt.ylabel('Total Area per 1000 ha')
plt.show()
```

## 2. Yearly Production
- Analysis: Identify the year with the highest rice production.
- Visualization: Line Chart
- How: Plot a line chart with years on the x-axis and rice production on the y-axis. Highlight the peak year with a different color or marker.



```python
# Identify the year with the highest rice production.
max_rice_pd = df.groupby(by = 'Year').agg({'RICE PRODUCTION (1000 tons)' : 'sum'}).reset_index()
max_rice_pd
```

```python
year = max_rice_pd.loc[max_rice_pd['RICE PRODUCTION (1000 tons)'] == max_rice_pd['RICE PRODUCTION (1000 tons)'].max() , 'Year']
year
```

```python
print(f"The year with the highest rice production is {year.values[0]}, with prodcution {max_rice_pd['RICE PRODUCTION (1000 tons)'].max()} per 1000 tons.")
```

```python
# Plot a line chart with years on the x-axis and rice production on the y-axis. 
# Highlight the peak year with a different color or marker.
sns.lineplot(x = max_rice_pd['Year'] , y = max_rice_pd['RICE PRODUCTION (1000 tons)'])
plt.title('Line Chart depicting Rice Production for particular Year.')
plt.show()
```

## 3. State Production
- Analysis: Find the state with the highest and lowest wheat production.
- Visualization: Horizontal Bar Chart
- How: Create a horizontal bar chart showing wheat production by state. This allows for easy comparison of states' production levels.


```python
state_wheat = df.groupby(by = 'State Name').agg({'WHEAT PRODUCTION (1000 tons)' : 'sum'}).reset_index()
state_wheat.sort_values(by = 'WHEAT PRODUCTION (1000 tons)',ascending=False , inplace=True)
state_wheat
```

```python
max_state = state_wheat.loc[state_wheat['WHEAT PRODUCTION (1000 tons)']== state_wheat['WHEAT PRODUCTION (1000 tons)'].max() , 'State Name']
min_state = state_wheat.loc[state_wheat['WHEAT PRODUCTION (1000 tons)']== state_wheat['WHEAT PRODUCTION (1000 tons)'].min() , 'State Name']
print(max_state)
print(min_state)
```

```python
print(f"The Satte with highest wheat production is {max_state.values[0]} and lowest wheat production is {min_state.values[0]} .")
```

```python
# Create a horizontal bar chart showing wheat production by state. 
# This allows for easy comparison of states' production levels.
plt.figure(figsize=(10,6))
sns.barplot(x = state_wheat['State Name'] , y = state_wheat['WHEAT PRODUCTION (1000 tons)' ])
plt.title('Bar chart showing wheat production by state')
plt.xticks(rotation = 45 , ha  = 'right' )
plt.show()
```

## 4. Crop Yields
- Analysis: Calculate the average yield for sorghum.
- Visualization: Box Plot
- How: Use a box plot to display yield distribution. This helps identify the median, quartiles, and any potential outliers in sorghum yields.

```python
print(f"The average yield for sorghum is {round(df['SORGHUM YIELD (Kg per ha)'].mean(),2)} Kg per ha.")
```

```python
plt.boxplot(df['SORGHUM YIELD (Kg per ha)'])
plt.show()
Q1 = df['SORGHUM YIELD (Kg per ha)'].quantile(0.25)
Q2 = df['SORGHUM YIELD (Kg per ha)'].quantile(0.5)
Q3 = df['SORGHUM YIELD (Kg per ha)'].quantile(0.75)
IQR = Q3 - Q1
LW = Q1 - (1.5*IQR)
UW = Q3 + (1.5*IQR)
outliers = ((df['SORGHUM YIELD (Kg per ha)']<LW) | (df['SORGHUM YIELD (Kg per ha)']>UW))
```

```python
print(f"The sorghum yields has \n Q1 :{Q1},\n Q2 :{Q2:.2f},\n Q3 :{Q3:.2f},\n IQR : {IQR:.2f},\n UW : {UW:.2f}, \n LW : {LW:.2f} and \n No. of Outliers : {outliers.sum()}")
```
